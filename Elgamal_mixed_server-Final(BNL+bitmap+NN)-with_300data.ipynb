```python
def fisher_yates_shuffle(the_list):
    list_range = range(0, len(the_list))
    for i in list_range:
        j = randint(list_range[0], list_range[-1])
        the_list[i], the_list[j] = the_list[j], the_list[i]
```

```python
import random
def shuffle_new(data):
    rand=[]
    for i in range(len(data)):
        rand.append(random.random())
    rand1=[]
    for i in range (len(data)):
        rand1.append(rand[i])
    rand.sort()
    
    index_rand=[]
    for i in range (len(data)):
        index_rand.append(rand1.index(rand[i]))
    area=[]
    price=[]
    for i in range (len(data)):
        area.append(data[i][0])
        price.append(data[i][1])
    for i in range (len(data)):
        data[i][0]=area[index_rand[i]]
        data[i][1]=price[index_rand[i]]
    return data
```
